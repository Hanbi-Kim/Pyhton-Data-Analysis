# 비디오게임 시장 변화 분석 1980년~2016년

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("dataset/vgsales.csv")
df
```

```python
df.describe()
```

# 매년 가장 전세계적으로 가장 많이 팔린 비디오게임 리스트화

```python
df.loc[df.groupby(['Year'])['Global_Sales'].idxmax()]
```

## 2017/2020년은 상대적으로 관측값이 작음으로 제거

```python
df=df.drop(df[df["Year"]==2020].index)
df=df.drop(df[df["Year"]==2017].index)
```

## 매해 출판한 비디오게임의 개수

```python
df.groupby("Year").count()
```

## 장르별 출판된 비디오게임 개수

```python
df_gr = df.groupby("Genre").count()
df_gr=df_gr.reset_index()
```

```python
x = df_gr["Genre"]
y = df_gr["Rank"]

plt.figure(figsize=(16, 7)) ## 그래프사이즈설정 
colors = sns.color_palette('hls',len(x)) ##씨본에서 제공하는 색팔레트 

plt.bar(x,y,color=colors)

plt.title('1980-2016년 장르별 비디오게임 판매량',fontsize=20) ## 타이틀 출력
plt.xlabel('비디오게임 장르') ## x축 라벨 출력
plt.ylabel('글로벌 비디오게임 판매량') ## y축 라벨 출력
plt.rc('font', family='NanumGothic') ## 한글깨짐 폰트 변경

plt.show()
```
